```python
import seaborn as sns
```

```python
unique_statuses = df['Status'].unique()
print(unique_statuses)
```

```python
total_days = df['Date'].nunique()
print(f"Total Days:{total_days}")
```

```python
present_days = df[df['Status'].isin(['P', 'MIS', 'Half Day'])].shape[0]
print(f"Total Present Days: {present_days}")
```

```python
absent_days = df[df['Status'] == 'A'].shape[0]
print(f"Total Absent Days: {absent_days}")
```

```python
df['Hours Worked']=df['Hours Worked'].astype('int')
```

```python
print(df['Hours Worked'].unique())
df['Hours Worked'] = pd.to_numeric(df['Hours Worked'], errors='coerce')
print(df[df['Hours Worked'].isna()])
df['Hours Worked'].fillna(0,inplace=True)
total_hours_worked = df['Hours Worked'].sum()
print(f"Total Hours Worked: {total_hours_worked}")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Status', data=df)
plt.title('Attendance Status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
employee_hours = df.groupby('Name')['Hours Worked'].sum().reset_index()
sns.barplot(x='Name', y='Hours Worked', data=employee_hours)
plt.title('Total Hours Worked by Employee')
plt.xlabel('Name')
plt.ylabel('Total Hours Worked')
plt.xticks(rotation=90)
plt.show()
```
